```python
cd = '/home/DAVIDSON/iarolls/Workspace/CSC371/HW3/data-3'
df = pd.DataFrame(columns = ['cancer_type', 'patient_id'])
dictionary = {'cancer_type':[],
              'patient_id': []
             }
id_dict = {'Breast Invasive Carcinoma': 0, 'Kidney Renal Clear Cell Carcinoma': 1, 'Lung Adenocarcinoma': 2, 'Lung Squamous Cell Carcinoma': 3, 'Pancreatic Adenocarcinoma': 4, 'Uveal Melanoma': 5}
cancer = 0


for cancerType in os.listdir(cd):
    cancer = id_dict[cancerType]
    newpath = os.path.join(cd, cancerType)
    for foldername in os.listdir(newpath):
        folder = os.path.join(newpath, foldername)
        if os.path.isdir(folder):
            for filename in os.listdir(folder):
                if filename=='annotations.txt' or filename=='MANIFEST-checkpoint.txt':
                    continue
                file = os.path.join(folder, filename)
                if os.path.isdir(file):
                    continue
                dictionary['patient_id'].append(foldername)
                dictionary['cancer_type'].append(cancer)
                df_ex = pd.read_csv(file, sep='\t', lineterminator='\n')
                for item in df_ex.index:
                    key = df_ex['miRNA_ID'][item]
                    if key not in dictionary.keys():
                        dictionary[key] = []
                    dictionary[key].append(df_ex['reads_per_million_miRNA_mapped'][item])


dataframe = pd.DataFrame.from_dict(dictionary)                
dataframe.head()
```

```python
dataframe.to_csv('/home/DAVIDSON/iarolls/Workspace/CSC371/HW3/cancerDataframe.csv', index = False)
```
