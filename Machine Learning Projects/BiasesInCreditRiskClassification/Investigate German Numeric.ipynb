```python
from sklearn.preprocessing import LabelEncoder

# assume your dataframe is called 'df'
categorical_cols = ['0', '2', '3', '5', '6', '8', '9', '11', '13', '14', '16', '18', '19']
numeric_cols = ['1', '4', '7', '10', '12', '15', '17']

# perform label encoding on categorical columns
le = LabelEncoder()
for col in categorical_cols:
    data_cat[col] = le.fit_transform(data_cat[col])
```

### Scale Data

```python
# get a summary of statistics for all numeric columns
summary = data_cat.describe()

# transpose the summary dataframe to switch rows and columns
summary = summary.transpose()

# select the mean, min, and max columns
#summary = summary[['mean', 'min', 'max']]
summary_pd = pd.DataFrame(summary)

# print the summary dataframe
summary_pd
```

```python
summary_pd['25%'][1]
```

```python
print(data_cat.isna().any())
```

```python
#if<25% -> 0, if <50% -> 1, if <75% -> 2, else 3
#1
thresholds1 = [summary_pd['min'][1]-1,summary_pd['25%'][1], summary_pd['50%'][1], summary_pd['75%'][1], summary_pd['max'][1]+1]

# Use cut() to create the categorical column
data_cat['1'] = pd.cut(data_cat['1'], bins=thresholds1, labels=[0, 1, 2, 3])
#4
thresholds4 = [summary_pd['min'][4]-1,summary_pd['25%'][4], summary_pd['50%'][4], summary_pd['75%'][4], summary_pd['max'][4]+1]
data_cat['4'] = pd.cut(data_cat['4'], bins=thresholds4, labels=[0, 1, 2, 3])
#12
thresholds12 = [summary_pd['min'][12]-1,summary_pd['25%'][12], summary_pd['50%'][12], summary_pd['75%'][12], summary_pd['max'][12]+1]
data_cat['12'] = pd.cut(data_cat['12'], bins=thresholds12, labels=[0, 1, 2, 3])
```
