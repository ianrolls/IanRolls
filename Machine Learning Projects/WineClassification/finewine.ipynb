```python
#Creating Colums for Squared and Cubed
df_red["fa2"]=df_red["fixed acidity"] ** 2
df_red["fa3"]=df_red["fixed acidity"] ** 3
df_red["lnfa"] = np.log(df_red["fixed acidity"])

df_red["va2"]=df_red["volatile acidity"] ** 2
df_red["va3"]=df_red["volatile acidity"] ** 3
df_red["lnva"] = np.log(df_red["volatile acidity"])

df_red["ca2"]=df_red["citric acid"] ** 2
df_red["ca3"]=df_red["citric acid"] ** 3

df_red["c2"]=df_red["chlorides"] ** 2
df_red["c3"]=df_red["chlorides"] ** 3
df_red["lnc"] = np.log(df_red["chlorides"])

df_red["s2"]=df_red["sulphates"] ** 2
df_red["s3"]=df_red["sulphates"] ** 3
df_red["lns"] = np.log(df_red["sulphates"])

df_red["a2"]=df_red["alcohol"] ** 2
df_red["a3"]=df_red["alcohol"] ** 3
df_red["lna"] = np.log(df_red["alcohol"])

#Creating Colums for Squared and Cubed
# df_white["fa2"]=df_white["fixed acidity"] ** 2
# df_white["fa3"]=df_white["fixed acidity"] ** 3
# df_white["lnfa"] = np.log(df_white["fixed acidity"])

# df_white["va2"]=df_white["volatile acidity"] ** 2
# df_white["va3"]=df_white["volatile acidity"] ** 3
# df_white["lnva"] = np.log(df_white["volatile acidity"])

# df_white["tsd2"]=df_white["total sulphur dioxide"] ** 2
# df_white["tsd3"]=df_white["total sulphur dioxide"] ** 3
# df_white["lntsd"] = np.log(df_white["total sulphur dioxide"])

# df_white["c2"]=df_white["chlorides"] ** 2
# df_white["c3"]=df_white["chlorides"] ** 3
# df_white["lnc"] = np.log(df_white["chlorides"])

# df_white["d2"]=df_white["density"] ** 2
# df_white["d3"]=df_white["density"] ** 3
# df_white["lnd"] = np.log(df_white["density"])

# df_white["a2"]=df_white["alcohol"] ** 2
# df_white["a3"]=df_white["alcohol"] ** 3
# df_white["lna"] = np.log(df_white["alcohol"])

col_at_end = ["quality"]
#Moves quality to end of DF
df_red = df_red[[c for c in df_red if c not in col_at_end] + [c for c in col_at_end if c in df_red]]
#print(df_red.head())
print(df_red.shape)


#Sklearn Scaling
target = df_red['quality']
data = df_red.iloc[:,0:28]
print(data.head())
scaler = MinMaxScaler()
scaler.fit(data)
data = scaler.transform(data)

# target = df_white['quality']
# data = df_white.iloc[:,0:29]
# scaler = MinMaxScaler()
# scaler.fit(data)
# data = scaler.transform(data)

df_red.columns.get_loc("fa2")
```

```python
from sklearn.model_selection import train_test_split, cross_validate, cross_val_score


(x_train, x_test, y_train, y_test) = train_test_split(data, target, test_size=0.2, random_state=1)
```

```python
#dummy model
from sklearn.dummy import DummyClassifier
reg_dummy = DummyClassifier(strategy = "most_frequent")

scores = cross_val_score(reg_dummy, x_train, y_train, cv=5, scoring='neg_mean_squared_error', error_score="raise")
print('CV score:', scores.mean(), "+/-", scores.std())

```

```python
#ridge model
from sklearn import linear_model
from sklearn.metrics import get_scorer_names
reg=linear_model.Ridge(alpha=.002, solver = "sag")
#reg.fit(x_train, y_train)

cv = cross_validate(reg, x_train, y_train, cv=5, scoring=('neg_mean_squared_error', 'r2'), error_score="raise")
print(cv.keys())
print('CV score:', cv['test_neg_mean_squared_error'].mean(), "+/-", cv['test_neg_mean_squared_error'].std())
```

```python
#Ridge CV Model
from sklearn.linear_model import RidgeCV

ridgeReg = RidgeCV(alphas=[1e-3, 2e-3, 1.5e-3, 3e-3, 1e-2], cv=10, scoring='neg_mean_squared_error').fit(x_train, y_train)
print(ridgeReg.score(x_train, y_train))
print(ridgeReg.alpha_)
print(ridgeReg.coef_)
print(ridgeReg.intercept_)

x = data[:,0]
y = df_red["quality"]
plt.figure(figsize=(4, 3))
ax = plt.axes()
ax.scatter(x, y, color='blue')
ax.plot(x, ridgeReg.intercept_ + ridgeReg.coef_[0] * x + ridgeReg.coef_[11] * (x**2) + ridgeReg.coef_[12]* (x**3), color='red')

ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()
```

```python
#lasso Model
reg_lasso=linear_model.Lasso(alpha=.01)
reg_lasso.fit(x_train, y_train)

#scorer = make_scorer(accuracy_score)
scores = cross_val_score(reg_lasso, x_train, y_train, cv=10, scoring = "neg_mean_squared_error", error_score="raise")
print('CV score:', np.mean(abs(scores)), "+/-", scores.std())
```
